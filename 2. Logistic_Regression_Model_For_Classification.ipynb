```python
import numpy as np
import math, copy
X_train = np.array([[0.5,1.5],[1,1],[1.5,0.5],[3,0.5],[2,2],[1,2.5]])
y_train = np.array([0,0,0,1,1,1])
w_tmp = np.array([1,1])
b_tmp = -3
def sigmoid(Z):

    g = 1 / (1 + np.exp(-Z))
    return g 

def compute_cost_logistic(X,y,w,b):

    m = X.shape[0]
    cost = 0
    for i in range(m):
        z_i = np.dot(X[i], w) + b 
        f_wb_i = sigmoid(z_i)
        cost += -y[i]*np.log(f_wb_i) - (1-y[i])*np.log(1-f_wb_i)
    cost = cost/m 
    return cost 
Optimal_cost = compute_cost_logistic(X_train,y_train,w_tmp,b_tmp)
print(Optimal_cost)
b = -4 
Optimal_cost = compute_cost_logistic(X_train,y_train,w_tmp,b)
print(Optimal_cost)

def compute_gradient_logistic(X,y,w,b):

    m,n = X.shape
    dj_dw = np.zeros((n,))
    dj_db = 0 
    for i in range(m):
        f_wb_i = sigmoid((np.dot(X[i], w )) + b )
        err_i = f_wb_i - y[i]
        for j in range(n):
            dj_dw[j] = dj_dw[j] + err_i * X[i,j]
        dj_db = dj_db + err_i 

    dj_dw = dj_dw / m 
    dj_db = dj_db / m 

    return dj_dw,dj_db 

w_tmp = np.array([2,3])
b_tmp = 1
dj_dw_tmp,dj_db_tmp = compute_gradient_logistic(X_train,y_train,w_tmp,b_tmp)

print(dj_dw_tmp,dj_db_tmp)

def compute_gradient_descent_logistic(X,y,w_in,b_in,num_iters,alpha):

    j_history = []
    w = copy.deepcopy(w_in)
    b = b_in


    for i in range(num_iters):
        dj_dw,dj_db = compute_gradient_logistic(X,y,w,b)
        w = w-alpha*dj_dw 
        b = b-alpha*dj_db
        if i <100000:
            j_history.append(compute_cost_logistic(X,y,w,b))
        if i % math.ceil(num_iters/10) == 0:
            print(f"Iterations: {i:4}, Cost: {j_history[-1]}")
    return w,b,j_history 

initial_w = np.zeros_like(X_train[0])
initial_b = 0
alpha = 0.1
iterations = 10000
w_final,b_final,_= compute_gradient_descent_logistic(X_train,y_train,initial_w,initial_b,iterations,alpha)
print(w_final,b_final)


m,_= X_train.shape
Predictions = np.zeros(m)
for i in range(m):
    Predictions[i] =  sigmoid(np.dot(X_train[i],w_final) + b_final)

print(Predictions)

yhat = np.zeros_like(Predictions)
for j in range(len(Predictions)):
    if Predictions[j]>=0.5:
        yhat[j]= 1
    else:
        yhat[j] = 0

print(yhat)
```
